```
def get_content(textfilepath):
  path=textfilepath
  #filename="abc.txt"
  #txtfile="juhi and meet".
  #path="/content/drive/My Drive/Projects/Characters/abc.txt"
  txtfile = open(path, "r")
  str1=txtfile.read()
  return str1
```

```
def make_images(text_data):
  str1=text_data
  img1=cv2.imread("/content/drive/My Drive/Projects/Characters/space.jpg")
  for i in str1:
    #print(i)
    if(i==" "):
      i="space"
    elif(i=="."):
      i="full_stop"
    elif(i==","):
      i="coma"
    elif(i==":"):
      i="colon"
    elif(i=="}"):
      i="closed_curly_bracket"
    elif(i==")"):
      i="closed_round_bracket"
    elif(i=="]"):
      i="closed_square_bracket"
    elif(i=="/"):
      i="divide"
    elif(i=="$"):
      i="dollar"
    elif(ascii(i)==34):#check here
      i="double_quote"
    elif(i=="="):
      i="equals"
    elif(i=="!"):
      i="exclamation"
    elif(i==">"):
      i="greater_than"
    elif(i=="#"):
      i="hash"
    elif(i=="@"):
      i="at_the_rate"
    elif(i=="-"):
      i="hyphen"
    elif(i=="<"):
      i="less_than"
    elif(i=="("):
      i="open_round_bracket"
    elif(i=="["):
      i="open_square_bracket"
    elif(i=="%"):
      i="percent"
    elif(i==";"):
      i="semicolon"
    elif(ascii(i)==39):
      i="single_quote"
    elif(ascii(i)==134):
      i="not_divide"
    elif(i=="^"):
      i="power"
    elif(i=="&"):
      i="ampercent"
    elif(i=="|"):
      i="abs"
    elif(i==0 or i==1 or i==2 or i==3 or i==4 or i==5 or i==6 or i==7 or i==8 or i==9):
      i=str(i)
    add="/content/drive/My Drive/Projects/Characters/"+i+".jpg"
    #print(add)
    img2=cv2.imread(add)
    img1=np.concatenate((img1, img2), axis=1)
  return img1
```
